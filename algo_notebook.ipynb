```python
import pandas as pd
import numpy as np
import talib

file = 'data/ReportTester-50760857-EURUSD-sell1.csv'
positions = pd.read_csv(file)
positions.dropna(inplace=True)

positions
```

```python
aux_positions_1 = pd.DataFrame()
aux_positions_2 = pd.DataFrame()

for row in positions.index:
    if row == 0:
        aux_positions_1 = aux_positions_1.append(positions.loc[row])
    elif (row % 2) == 0:
        aux_positions_1 = aux_positions_1.append(positions.loc[row])
    else:
        aux_positions_2 = aux_positions_2.append(positions.loc[row])
```

```python
aux_positions_1.reset_index(drop=True, inplace=True)
aux_positions_2.reset_index(drop=True, inplace=True)
```

```python
positions = pd.concat([aux_positions_1, aux_positions_2], axis=1)
```

```python
my_columns = ['Entry Date', 'Entry Deal', 'Symbol', 'Entry Type', 
              'Entry Direction', 'Entry Volume', 'Entry Price', 'Entry Profit',
              'Exit Date', 'Exit Deal', 'Exit Symbol', 'Exit Type', 
              'Exit Direction','Exit Volume','Exit Price', 'Exit Profit']

positions.set_axis(my_columns, axis=1, inplace=True)
positions.drop(['Entry Profit', 'Exit Symbol'] , axis=1, inplace=True)
```

```python
positions['Entry Date'] = pd.to_datetime(positions['Entry Date'])
```

```python
positions
```

```python
file = 'data/EURUSD_H4_200601020000_202112302000.csv'
symbol_data = pd.read_csv(file, sep='\t')

symbol_data
```

```python
symbol_data['Date'] = symbol_data['<DATE>'] + ' ' + symbol_data['<TIME>']
col = symbol_data.pop('Date')
symbol_data.insert(0, "Date", col)
symbol_data.drop(['<DATE>','<TIME>'], axis=1, inplace=True)
```

```python
atr_symbol_data = talib.ATR(symbol_data['<HIGH>'].values, symbol_data['<LOW>'].values, 
                     symbol_data['<CLOSE>'].values, timeperiod=14)
```

```python
symbol_data['ATR'] = atr_symbol_data
```

```python
slowk, slowd = talib.STOCH(symbol_data['<HIGH>'].values, 
                           symbol_data['<LOW>'].values, 
                           symbol_data['<CLOSE>'].values,
                           fastk_period=14, slowk_period=3, slowk_matype=0, 
                           slowd_period=3, slowd_matype=0)
slowk

symbol_data['STOCH slowK'] = slowk
```

```python
aroondown, aroonup = talib.AROON(symbol_data['<HIGH>'].values, 
                                 symbol_data['<LOW>'].values, 
                                 timeperiod=25)

symbol_data['AROON down'] = aroondown
symbol_data['AROON up'] = aroonup
```

```python
symbol_data['Date'] = pd.to_datetime(symbol_data['Date'])
symbol_data
```

```python
pos_atr = np.array([])
pos_stoch_slowk = np.array([])
pos_aroon_down = np.array([])
pos_aroon_up = np.array([])
pos_timeframe_date = np.array([])
row_symbol_data = 0
interval = pd.Timedelta(seconds=1)

for row_positions in positions.index:
    
    for row_symbol_data in range(row_symbol_data, len(symbol_data)+1):

        if row_symbol_data == 0:
            continue
        
        else:
            pos_signal_date = positions.loc[row_positions, 'Entry Date'] - interval
            previous_symbol_date = symbol_data.loc[row_symbol_data - 1, 'Date']
            now_symbol_date = symbol_data.loc[row_symbol_data, 'Date']
        
            if pos_signal_date >= previous_symbol_date and \
                pos_signal_date < now_symbol_date:                
                pos_atr = np.append(pos_atr, 
                                      symbol_data.loc[row_symbol_data-1, 
                                                      'ATR'])
                pos_stoch_slowk = np.append(pos_stoch_slowk, 
                                            symbol_data.loc[row_symbol_data-1,
                                                            'STOCH slowK'])
                pos_aroon_down = np.append(pos_aroon_down,
                                           symbol_data.loc[row_symbol_data-1,
                                                           'AROON down'])
                pos_aroon_up = np.append(pos_aroon_up,
                                         symbol_data.loc[row_symbol_data-1,
                                                         'AROON up'])
                pos_timeframe_date = np.append(pos_timeframe_date,
                                        symbol_data.loc[row_symbol_data-1,
                                                        'Date'])    
                row_symbol_data -= 1            
                break
            
positions['ATR'] = pos_atr
positions['STOCH slowK'] = pos_stoch_slowk
positions['AROON down'] = pos_aroon_down
positions['AROON up'] = pos_aroon_up
positions['TF Date'] = pos_timeframe_date
```

```python
positions
```

```python
positions.dropna(inplace=True)
positions.reset_index(drop=True, inplace=True)
s_stoch_slowk = np.array([])
s_aroon_down = np.array([])
s_aroon_up = np.array([])

for row in positions.index:
    stoch_slowk = positions.loc[row, 'STOCH slowK']
    aroondown = positions.loc[row, 'AROON down']
    aroonup = positions.loc[row, 'AROON up']

    if stoch_slowk > 70:
        s_stoch_slowk = np.append(s_stoch_slowk, 1)
    elif stoch_slowk < 30:
        s_stoch_slowk = np.append(s_stoch_slowk, -1)
    else:
        s_stoch_slowk = np.append(s_stoch_slowk, 0)
        
    if aroondown > 70:
        s_aroon_down = np.append(s_aroon_down, -1)
    else:
        s_aroon_down = np.append(s_aroon_down, 0)
        
    if aroonup > 70:
        s_aroon_up = np.append(s_aroon_up, 1)
    else:
        s_aroon_up = np.append(s_aroon_up, 0)
        
positions['Sentiment STOCH slowK'] = s_stoch_slowk
positions['Sentiment AROON down'] = s_aroon_down
positions['Sentiment AROON up'] = s_aroon_up
        
```

```python
positions
```

```python
positions['Exit Profit'] = positions['Exit Profit'].astype(float)
```

```python
mean_losses = positions['Exit Profit'][(positions['Exit Profit'] < 0)].mean()
abs(mean_losses)
positions['R'] = positions['Exit Profit'] / abs(mean_losses)
mean_losses
```

```python
mean_ATR = positions['ATR'].mean()
mean_ATR
```

```python
std_ATR = positions['ATR'].std()
std_ATR
```

```python
upper_ATR = mean_ATR + std_ATR
lower_ATR = mean_ATR - std_ATR

for row in positions.index:
    if positions.loc[row,  'ATR'] >= upper_ATR:
        positions.loc[row, 'Sentiment ATR'] = 'High Volatility'
    elif positions.loc[row,  'ATR'] <= lower_ATR:
        positions.loc[row, 'Sentiment ATR'] = 'Low Volatility'
    else:
        positions.loc[row, 'Sentiment ATR'] = 'Medium Volatility'
        
positions
```
